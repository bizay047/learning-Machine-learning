```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
```

```python
iris = load_iris()

X = iris.data
y = iris.target
```

```python
feature_names = iris.feature_names
target_names = iris.target_names

feature_names, target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
plt.figure(figsize=(16, 10))
plot_tree(
    model,
    feature_names=feature_names,
    class_names=target_names,
    filled=True
)
plt.title("Decision Tree Visualization (Iris Dataset)")
plt.show()
```

```python
print(
    "The Decision Tree classifier effectively learned non-linear decision boundaries "
    "and classified Iris species with high accuracy."
)
print(
    "In this experiment, a Decision Tree classifier was developed to classify flower species "
    "in the Iris dataset. The model achieved good accuracy and was evaluated using precision, "
    "recall, and F1-score. Visualization of the decision tree provided clear insight into how "
    "the model splits data at different nodes."
)

```
